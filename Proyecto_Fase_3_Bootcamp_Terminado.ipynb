```python
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

ticker = sp500 ['Symbol']

all_data = []
for tickers in ticker:
  data = yf.download(tickers, start='2024-04-01', end='2024-06-28')
  data['ticker'] = tickers
  all_data.append(data)
  combined_data = pd.concat(all_data)

def extract_data(ticker, start_date, end_date):
    try:
        logging.info(f'Extrayendo datos para {ticker} desde {start_date} hasta {end_date}')
        data = yf.download(ticker, start=start_date, end=end_date)
        spdata= data.append()
        logging.info(f'Datos extrados exitosamente para {ticker}')
        return data
    except Exception as e:
        logging.error(f'Error extrayendo datos para {ticker}: {e}')
        return None
```
